###Description:
####Calculate and visually represent the overall churn rate. Explore customer distribution by gender, partner status, and dependent status. Analyze tenure distribution and its relation with churn. Investigate how churn variesacross different contract types and payment methods.

```python
# Importing Libraries

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load dataset
df = pd.read_csv("/content/Telco_Customer_Churn_Dataset.csv")
```

```python
# Fix TotalCharges formatting
df["TotalCharges"] = df["TotalCharges"].replace(" ", pd.NA)
df["TotalCharges"] = pd.to_numeric(df["TotalCharges"], errors="coerce")
```

```python
# 1. Overall Churn Rate
churn_rate = df["Churn"].value_counts(normalize=True) * 100
print("Churn Rate (%):\n", churn_rate)

plt.figure(figsize=(5,4))
sns.countplot(data=df, x="Churn", palette="viridis")
plt.title("Overall Churn Count")
plt.show()
```

```python
# 2. Distribution by Gender
plt.figure(figsize=(6,4))
sns.countplot(data=df, x="gender", hue="Churn", palette="viridis")
plt.title("Churn by Gender")
plt.show()
```

```python
# 3. Distribution by Partner

plt.figure(figsize=(6,4))
sns.countplot(data=df, x="Partner", hue="Churn", palette="viridis")
plt.title("Churn by Partner Status")
plt.show()
```

```python
# 4. Distribution by Dependents

plt.figure(figsize=(6,4))
sns.countplot(data=df, x="Dependents", hue="Churn", palette="viridis")
plt.title("Churn by Dependents Status")
plt.show()
```

```python
# 5. Tenure Distribution

plt.figure(figsize=(7,4))
sns.histplot(df["tenure"], kde=True, bins=30)
plt.title("Tenure Distribution")
plt.xlabel("Tenure (Months)")
plt.show()
```

```python
# Churn vs Tenure
plt.figure(figsize=(7,4))
sns.boxplot(data=df, x="Churn", y="tenure", palette="viridis")
plt.title("Tenure vs Churn")
plt.show()
```

```python
# 6. Churn across Contract Types

plt.figure(figsize=(7,4))
sns.countplot(data=df, x="Contract", hue="Churn", palette="viridis")
plt.title("Churn Across Contract Types")
plt.xticks(rotation=20)
plt.show()
```

```python
# 7. Churn Across Payment Methods

plt.figure(figsize=(8,4))
sns.countplot(data=df, x="PaymentMethod", hue="Churn", palette="viridis")
plt.title("Churn Across Payment Methods")
plt.xticks(rotation=20)
plt.show()
```
