# Task 6 : Business Recommendations

###Description:
#### Based on the analysis and predictive models, provide actionable recommendations to the business. Suggest specific marketing strategies, retention offers, or customer engagement tactics. Estimate the potential impact of these recommendations on revenue and churn rate

# 1. Overview

Based on the insights from EDA, segmentation, and the prediction model, several customer groups show a significantly higher risk of churn. These insights help form targeted actions that improve retention and reduce revenue loss.

# 2. Key Factors Driving Churn

From the model’s feature importance and earlier analysis, churn is strongly influenced by:

    i.) Month-to-month contracts

    ii.) High monthly charges

    iii.) Low tenure (0–12 months)

    iv.) Lack of security, tech support, and online backup services

    v.) Electronic check payment method

These areas should be the focus of business strategy.

# 3. Actionable Recommendations
#### A. Improve Retention for Month-to-Month Customers

These customers had the highest churn rate in segmentation.

    i.) Offer discounted upgrades to 1-year or 2-year contracts.

    ii.) Provide limited-time loyalty perks for customers completing 6 months.

    iii.) Introduce “stay benefits” such as free add-ons for the first 3 months.

Expected Impact: Lower churn among the largest at-risk group and more predictable long-term revenue.

#### B. Reduce Churn Among High-Value Customers

High-value but at-risk customers were identified in Task 3.

    i.) Provide personalized retention offers such as small bill reductions or bundled services.

    ii.) Assign priority customer support.

    iii.)Offer early-renewal incentives before contract expiry.

Expected Impact: Saves significant revenue because high-value users generate more ARPU.

#### C. Lower Monthly Charges for High-Risk Segments

Customers with high or very high monthly charges churn more often.

    i.) Introduce custom plans based on actual usage.

    ii.) Provide first-time bill reductions or loyalty credits.

    iii.) Offer flexible add-on packages instead of fixed bundles.

Expected Impact: Lowers dissatisfaction and improves long-term customer value.

#### D. Improve Service Adoption (Security, Tech Support, Backup)

Model analysis showed that customers not using these services churn more.

    i.) Create low-cost bundles combining security + tech support.

    ii.) Offer a free 30-day trial for any add-on service.

    iii.) Use email/SMS to highlight benefits (protection, faster support, backup safety).

Expected Impact: Customers with more services tend to stay longer, reducing churn.

#### E. Reduce Issues with Electronic Check Payments

This payment method had the highest churn rate.

    i.) Encourage customers to switch to auto-payment or credit card by offering cashback.

    ii.) Provide reminders before due dates to avoid payment-related dissatisfaction.

Expected Impact: Fewer payment failures and improved customer experience.

# 4. Estimated Impact of Recommendations

    i.) Churn reduction: 8–15% depending on implementation depth

    ii.) Increase in contract renewals: 10–20% from long-term plan upgrades

    iii.) Revenue stability: Higher due to more customers on yearly contracts

    iv.) Lower support costs: As customers adopt tech support and security add-ons

    v.) Customer satisfaction improvement: From targeted offers and personalized retention

# 5. Final Summary

The combined insights from Tasks 1–5 show that churn is largely predictable and preventable.
By focusing on contract upgrades, reducing high charges, promoting service add-ons, and improving customer engagement, the business can significantly reduce churn and improve long-term profitability.
